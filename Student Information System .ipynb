```python
stu_info={}


def addStudent():
    rno =int(input("Enter Student Rollno"))
    name=input("ENter Student Name:")
    contact=int(input("Enter Student Contact Number"))
    dept =input("Enter Departemnt")
    add = input("Enter Student Address")
    
    stu_info[rno]=[name,contact,dept,add]
    
    print("\n Successfully Added Record")
    

def getStudentInfo():
    rno=int(input("Whose details U need"))
    if rno in stu_info.keys():
        print(stu_info[rno])
    else:
        print("\nDetails are not exist")
        

def deleteStudent():
    rno=int(input("Whose details want to delete"))
    if rno in stu_info.keys():
        del stu_info[rno]
        print("\n Record Deleted successfully")
        
    else:
        print("\n deatils are not exist")
        

def updateInfo():
    rno = int(input("Whose deatils u need ot update"))
    if rno in stu_info.keys():
        print("1. Update contact \n 2. Update Dept \n 3. Update address")
        choice = int(input("Enter Your Choice"))
        if choice==1:
            stu_info[rno][1]=int(input("Enter New Contact Number"))
        elif choice ==2:
            stu_info[rno][2]=input("Enter New Deaprtemnet")
        elif choice==3:
            stu_info[rno][3]=input("ENter New Address")
        else:
            print("Wrong choice")        
        
    else:
        print("Wrong Deatils")
        



while True:
    print("\n\n\n 1. Add Student \n 2. getStudent Information \n 3. Update student informatin \n 4. Delete Student \n ")

    choice = int(input("Enter Your choice"))

    if choice==1:
        addStudent()
        
    elif choice==2:
        getStudentInfo()
    elif choice ==3:
        updateInfo()
    elif choice==4:
        deleteStudent()
    else:
        print("wrong choice")

    flag=input("if you want to continue press [Y/y]")

    if flag=='y' or flag=='Y':
        pass
    else:
        break


print("Task has completed")
```
